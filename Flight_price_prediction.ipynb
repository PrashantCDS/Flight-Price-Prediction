Flight ticket prices can be something hard to guess, today we might see a price, check out the price of the same flight tomorrow, it will be a different story. We might have often heard travelers saying that flight ticket prices are so unpredictable. That’s why we will try to use machine learning to solve this problem. # Attribute Information : 

1.	Airline: So this column will have all the types of airlines like Indigo, Jet Airways, Air India, and many more.  

2.	Date_of_Journey: This column will let us know about the date on which the passenger’s journey will start.

3.	Source: This column holds the name of the place from where the passenger’s journey will start.

4.	Destination: This column holds the name of the place to where passengers wanted to travel.

5.	Route: Here we can know about what the route is through which passengers have opted to travel from his/her source to       their destination.

6.	Arrival_Time: Arrival time is when the passenger will reach his/her destination.

7.	Duration: Duration is the whole period that a flight will take to complete its journey from source to destination.

8.	Total_Stops: This will let us know in how many places flights will stop there for the flight in the whole journey.

9.	Additional_Info: In this column, we will get information about food, kind of food, and other amenities.

10.	Price: Price of the flight for a complete journey including all the expenses before onboarding.

#### cleaning the data

```python
data["Journey_day"] = pd.to_datetime(data.Date_of_Journey,format="%d/%m/%Y").dt.day
data["Journey_month"] = pd.to_datetime(data.Date_of_Journey,format="%d/%m/%Y").dt.month
```

```python
data.head(1)
```

```python
data["Dep_hours"] = pd.to_datetime(data.Dep_Time).dt.hour
data["Dep_mins"] = pd.to_datetime(data.Dep_Time).dt.minute
```

```python
data["Arrival_hours"] = pd.to_datetime(data.Arrival_Time).dt.hour
data["Arrival_mins"] = pd.to_datetime(data.Arrival_Time).dt.minute
```

```python
duration = list(data["Duration"])#creating new duration list column
for i in range(len(duration)):
    if len(duration[i].split()) !=2: # checking if duration having only hours and minutes
        if "h" in duration[i]:
            duration[i] = duration[i].strip()+" 0m" #adding 0minutes
        else:
            duration[i] = "0h "+duration[i] #adding ohours
Duration_hours = []# create new variable for duration hours
Duration_mins = []# create new variable for duration minutes
for i in range(len(duration)):
    Duration_hours.append(int(duration[i].split(sep = "h")[0])) #get all hours value in Duration_hours
    Duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))#get all minutes value in Duration_mins
    
data["Duration_hours"] = Duration_hours  #adding Duration_hours columns in data
data["Duration_mins"] = Duration_mins  #adding Duration_mins columns in data
```

```python
data.head()
```

```python
#Now,we don't need Date_of_Journey,Dep_Time,Arrival_Time,Duration.So,we have to drop this columns
#We can see Route and Total_stops are giving information almost same.so we can drop anyone columns from both of them.
data.drop(["Date_of_Journey","Dep_Time","Arrival_Time","Duration","Route","Additional_Info"],axis=1,inplace=True)
```

### RandomForestRegressor with Hyperparameters Tunning

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start=200, stop=2000, num=10)]#List Comprehension-using for loop in list
max_features = ['auto', 'sqrt']#maximum number of features allowed to try in individual tree
max_depth = [int(x) for x in np.linspace(10, 110, num=11)]#List Comprehension-using for loop in list
max_depth.append(None)
min_samples_split = [2, 5, 10]#minimum number of samples required to split an internal node
min_samples_leaf = [1, 2, 4]#minimum number of samples required to be at a leaf node.
bootstrap = [True, False]#sampling 


#dictionary for hyperparameters
random_grid = {'n_estimators': n_estimators, 'max_features': max_features,
               'max_depth': max_depth, 'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf, 'bootstrap': bootstrap}


rf_clf1 = RandomForestRegressor(random_state=42)#model

rf_cv = RandomizedSearchCV(estimator=rf_clf1, scoring='r2',param_distributions=random_grid, n_iter=100, cv=3, 
                               verbose=2, random_state=42, n_jobs=-1)



rf_cv.fit(x_train, y_train)##training data on randomsearch cv
rf_best_params = rf_cv.best_params_##it will give you best parameters 
print(f"Best paramters: {rf_best_params})")##printing  best parameters
 
```

# Report on Challenges faced

* This is a __Flight Price prediction__ which has __10683__ entries, 0 to 10682 Data and total columns are __11__. 10         columns having __object datatype__ and 1 columns which is our target variable it has __integer datatype__ in continous.
* In our dataset only two columns having __1 missing value__ Route and Total Stops.So i simply fill missing value by help     of __measure of tendency(mean,median,mode)__.
* Additional_Info having __no information__ of more than __80%__ data.So this columns is not useful for our prediction       that's why i   droped this columns.
* In our dataset some columns having date and time but that's in form of __object datatype__ which are                       Date_of_Journey,Dep_Time,Arrival_Time,Duration. I just change their datatype by help of __timestamp__ for converting into   timestamp with seperately columns for minutes and hours,i called __pd.to_datetime__.after converting those columns         seperately   we don't need real columns so i dropped that's.
* Duration columns having some  value only in __minutes__ and some value only in __hours__ so for that i just run loop       which columns having only hours i added their 0m and which columns having only mins i just added 0h then create a           seperatly columns for hours and minutes then split hours and min and appended in seprete columns.after that we don't need   actual columns so i dropped it.
* some columns are __nominal__ so for that i used __OneHotEncoding and ManualEncoding__.which are                             Airline,Source,Destination and total stops.
* In feature Selection every columns are __co-related__ that's why i used all columns in feature selection.and used some     different __Regression Algorithmns__.
